```python
import numpy as np
import math
import matplotlib.pyplot as plt

if __name__ == '__main__':
  x = np.arange(-5, 5, 0.01) # creates an array of x values from -5 to 5 by 0.01 step
  y = np.power(x, 2) # raises x values by 2 to get y values
  z = np.sqrt(x) # returns the non-negative square-root of an array, shows an error `invalid value encountered in sqrt` when negative values are encountered
  plt.plot(x, y) # generates graph
  plt.plot(x, z) # generates graph
  plt.title("Pow and sqrt functions")
  plt.show() # from displaying the graph
```
